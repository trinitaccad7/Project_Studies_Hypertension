# Introductions

<p style='text-align: justify;'>Hypertension has been known as one of the leading preventable risk factor for cardiovascular disease and responsible for approximately 50% of cardiovascular morbidity and mortality worldwide (1). In addition to being a major risk factor for cardiovascular disease, it was also found to be a risk factor for renal and cerebrovascular diseases (2). The associated risks of being diagnosed with hypertension was due to arterial wall modifications (during ageing), and the increased exposure to unhealthy lifestyle, such as unhealthy diets and lack of physical activities. This means that the adult population were at higher risks to suffer with hypertension, where those that were diagnosed with the disease falls in majority to the category of primary (essential) hypertension while a few hypertensive patients have secondary hypertension. To add context to the differences of the types of hypertension, primary hypertension is characterized as high blood pressure without an identifiable cause while secondary hypertension is associated by known disease or condition (3). Moreover, in a general sense, hypertension is characterized with chronic elevation of blood pressure, which in the long run would cause organ damage and eventually lead to a cascade of systematic problems in the body. Therefore, as a general practice, the blood pressure level of 140/90mm Hg became an indicator where treatment of hypertension should begin to prevent chronic complications (3).

<p style='text-align: justify;'>For blood pressure regulation under physiological conditions, the Renin-Angiotensin Aldesteron system (RAAS) functions to regulate blood pressure and hydro-electrolytic balance. The RAAS system works by cleavage of the susbtrate Angeiotensinogen (AGT) into Angiotensin-I by the enzyme Renin, and then further cleaved by Angiotensin-Converting Enzyme (ACE) into Angiotensin-II, which was the key player in the system. With this biological knowledge of the RAAS sytem and its key functionality in the regulation of blood pressure, therapeutic drugs that functions as inhibitors of the enzymes that participate in the RAAS system, for instance, Renin inhibitors, ACE inhibitors, Angeiotensin-II type 1 receptor antagonists, and mineralocorticoid receptor antagonists, in turn would help regulate the disregulated RAAS system and treat/prevent the event of hypertesion in high-risk patients (4).


<p style='text-align: justify;'>In the field of genetics and genomics, the Framingham Heart Study had recorded blood pressure data for three generations and with the absence of environmental factors, it was shown that children whose parents and grandparents diagnosed with hypertension had an increased risk of having the condition (6). In addition, a monogenic hypertension research had found that 12 genes (CYP11B1, CYP11B2, WNK1, WNK4, KLHL3, CUL3, SCNN1B, SCNN1G, CYP17A1, HSD11B2, NR3C2 and KCNJ5) in which 8 of these genes were involved in Mendelian syndromes causing Hypertension. Furthermore, 5 out of the 12 monogenic hypertension genes had upregulated expressions mutation while the remaining genes had downregulation expression mutation that led the reduction of inhibitory effects on blood pressure. Moreover, the 12 genes in the study belonged to two groups of pathways, such as the renal sodium handling pathway and steroid hormone metabolism pathway (1). Another case control study of monogenic forms of hypertension, such that genes were identified to determine the key mechanisms in the pathophysiology aspect of hypertension, especially those involved in the RAAS, catecholamine/adrenergic system, renal kinin system, epithelial sodium channel, adducin, and including those involved in lipoprotein metabolism, hormone receptors, and growth factors. The list of genes found in this review include AGT variants, adregenic receptors (ADRA1B and ADRB2), dopamine D1 receptor (DRD1), AGTR1, REN, ACE, glucocorticoid-remediable aldosteronism (GRA), 11-β-hydroxylase (CYP11B1), aldosterone synthase (CYP11B2), epithelial sodium channel three subunits (SCNN1A, SCNN1B and SCNN1G) (6).

### ClinVar Database - Determine variants indentified from patients associated with disease Hypertension

- As an alternative to searching via python, the web browser was used to retrieve the target information.
- The series of steps performed was as follow:
 1. Go to https://www.ncbi.nlm.nih.gov/clinvar/
 2. Search the term 'Hypertension'
 3. Download Tab-Delimited File by 'Relevance'

```python
%%bash

# This script is to use Unix EntrezDirect in searching through the Clinvar database
# then xtract the data into a tab-delimited file

# But first, Add headers to the output file
echo -e 'Title\tGene(s)\tTrait_Name\tAccession\tChromosome\tVariant_Type' > /mnt/c/Users/trini/Desktop/clinvar_result.txt

echo 'Headers Added'

# Execute EntrezDirect and append the output data
echo 'Executing EntrezDirect'
esearch -db clinvar -query 'Hypertension OR High blood pressure' |\
esummary |\
xtract -pattern DocumentSummary -sep "|" -element title symbol trait_name accession chr variant_type >> /mnt/c/Users/trini/Desktop/clinvar_result.txt

echo 'Done'
```

#### Parse the Tab-Delimited File from EntrezDirect clinvar with only 'hypertension'

```python
with open("/Users/trini/Desktop/clinvar_result.txt") as file:
    out = open("/Users/trini/Desktop/clinvar_hypertension.txt",'w')
    for line in file:
        out.write(line)
        break  # This way we can parse the header
    for l in file:
        if 'hypertension' in l:
            out.write(l)
    out.close()
```

### Now Let's begin searching through OMIM Database via Web-browser

1. Go to https://www.ncbi.nlm.nih.gov/omim/
2. Search using the term "Hypertension" OR "high blood pressure"
3. Download the output table

```python
# Extract the gene name from the 'Title' header
omim_file = open("/Users/trini/Downloads/OMIM-Entry-Retrieval.txt",'r')
title_list = []
for line in omim_file:
    s = line.split('\t')
    title = s[1]
    if ';' in title:
        title_list.append(title)
#print(title_list)
output_file = open("/Users/trini/Downloads/OMIM_gene_list.txt",'w')
gene_list = []
for gene in title_list:
    s = gene.split(';')
    gene_name = s[1].replace('"','')
    gene_name1 = gene_name.replace(' ','')
    gene_list.append(gene_name1)
    output_file.write(gene_name1)
```

### Include only those potential gene targerts that have OMIM data results

```python
# Merge all the lists of potential genes
Kegg_ClinVar_OMIMList = ['NR3C2', 'BMPR2', 'SMAD9', 'CAV1', 'KCNK3', 'ATP13A3', 'ACVRL1', 'EIF2AK4', 'CYP11B2', 'CYP11B1', 'HSD11B2', 'SCNN1B', 'SCNN1G']
Literature1 = ['CYP11B1', 'CYP11B2', 'WNK1', 'WNK4', 'KLHL3', 'CUL3', 'SCNN1B', 'SCNN1G', 'CYP17A1', 'HSD11B2', 'NR3C2', 'KCNJ5']
Literature2 = ['AGT', 'DRD1', 'AGTR1', 'REN', 'ACE', 'SCNN1A', 'SCNN1G']
genes = []
for L1 in Kegg_ClinVar_OMIMList:
    if L1 not in genes:
        genes.append(L1)
for L2 in Literature1:
    if L2 not in genes:
        genes.append(L2)
for L3 in Literature2:
    if L3 not in genes:
        genes.append(L3)
        
#print(sorted(genes))

# Include only those potential genes that were available in the OMIM database
targets_biomarkers = []
for markers in genes:
    if markers in gene_list:
        targets_biomarkers.append(markers)
    else:
        continue
print(sorted(targets_biomarkers)) # Now we have gene_targets that have OMIM data, which would be important in ClinicalTrials.gov search
```
